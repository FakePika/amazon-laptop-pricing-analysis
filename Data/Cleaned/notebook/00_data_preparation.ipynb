```python
import pandas as pd

df = pd.read_excel("C:/Users/prash/Desktop/Amazon/Data/Cleaned/amazon-laptop-pricing-analysis.xlsx")


```

```python
# Drop Octoparse/non-analysis columns
df = df.drop(columns=["Frequency", "Frequency Percent", "Page_Title"], errors="ignore")
```

```python
# Renaming the column names 
df = df.rename(columns={
    "Company": "brand",
    "Title": "product_title",
    "Price": "price",
    "Rating": "rating",
    "Number_of_Reviews": "review_count",
    "Satisfied": "satisfied"
})
```

```python
# Keeping only the columns we actually want
keep = ["brand", "product_title", "price", "rating", "review_count", "satisfied"]
df = df[[c for c in keep if c in df.columns]]
```

```python
import numpy as np

# numeric conversions
for col in ["price", "rating", "review_count"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# satisfied as 0/1 if it exists
if "satisfied" in df.columns:
    # handles cases like "Yes/No", "True/False", 1/0, etc.
    df["satisfied"] = df["satisfied"].replace(
        {"Yes": 1, "No": 0, "yes": 1, "no": 0, True: 1, False: 0}
    )
    df["satisfied"] = pd.to_numeric(df["satisfied"], errors="coerce")

df.info()
```

```python
out_path = "C:/Users/prash/Desktop/Amazon/Data/Cleaned/amazon_laptops_cleaned_public.csv"
df.to_csv(out_path, index=False)

out_path
```
